## Getting started
Load the breast cancer dataset from scikit-learn. Split the dataset into features (X) and target (y). Split the dataset into training and test sets.



```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importing modules from scikit-learn
from sklearn import datasets, metrics, model_selection, ensemble
from sklearn.model_selection import train_test_split

# Load the breast cancer dataset from scikit-learn
breast_cancer = datasets.load_breast_cancer()

# Split the dataset into features (X) and target (y)
#X = data.drop(columns='target')
#y = data['target']
X = breast_cancer.data
y = breast_cancer.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Gradient boosting

Use the GradientBoostingClassifier function from sklearn to train a gradient boosting model for breast cancer classification.
Evaluate your model on the test set, producing a confusion matrix and a classification report.


```python
from sklearn.metrics import confusion_matrix, classification_report

# Initialize and train a Gradient Boosting Classifier
gb_classifier = ensemble.GradientBoostingClassifier(random_state=42)
gb_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = gb_classifier.predict(X_test)

# Evaluate the model: Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)

# Evaluate the model: Classification report

class_report = classification_report(y_test, y_pred)
print("\nClassification Report:\n", class_report)
```

## XG Boost

Use the xgboost library to train an XGBoost model for breast cancer classification.  Evaluate your model on the test set, producing a confusion matrix and a classification report.

```python
# Import necessary libraries
import xgboost as xgb
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Load the breast cancer dataset from scikit-learn
breast_cancer = datasets.load_breast_cancer()

# Convert the dataset into DataFrame for easier handling
X = breast_cancer.data
y = breast_cancer.target

# Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the XGBoost model
xgb_model = xgb.XGBClassifier(random_state=42)
xgb_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = xgb_model.predict(X_test)

# Evaluate the model: Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)

# Evaluate the model: Classification report
class_report = classification_report(y_test, y_pred)
print("Classification Report:\n", class_report)

# Calculate and print accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

## LightGBM

Use the lightgbm library to train a LightGBM model for breast cancer classification. Evaluate your model on the test set, producing a confusion matrix and a classification report.




```python
# Import necessary libraries
import lightgbm as lgb
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score


# Create and train the LightGBM model
lgb_model = lgb.LGBMClassifier(random_state=42)
lgb_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = lgb_model.predict(X_test)

# Evaluate the model: Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)

# Evaluate the model: Classification report
class_report = classification_report(y_test, y_pred)
print("Classification Report:\n", class_report)

# Calculate and print accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

```

## AdaBoost

Use the AdaBoostClassifier function from sklearn to train an Adaptive boosting model for breast cancer classification.
Evaluate your model on the test set, producing a confusion matrix and a classification report.


```python
# Import necessary libraries
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score


# Create and train the AdaBoost model
adaboost_model = AdaBoostClassifier(random_state=42, n_estimators=50)
adaboost_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = adaboost_model.predict(X_test)

# Evaluate the model: Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)

# Evaluate the model: Classification report
class_report = classification_report(y_test, y_pred)
print("Classification Report:\n", class_report)

# Calculate and print accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

## Hyperparameter tuning

Use the sklearn GridSearchCV to create a parameter grid for each model and find the best estimators for each.


```python
from sklearn.model_selection import GridSearchCV
import xgboost as xgb
import lightgbm as lgb
from sklearn.ensemble import AdaBoostClassifier

# Parameter grid for tuning (common across models for simplicity)
param_grid_estimators_rate = {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2]}

# ---- Tuning Gradient Boosting ----
grid_search_gb = GridSearchCV(estimator=ensemble.GradientBoostingClassifier(random_state=42),
                              param_grid=param_grid_estimators_rate, cv=3, scoring='accuracy')
grid_search_gb.fit(X_train, y_train)
best_gb = grid_search_gb.best_estimator_
print("Best Hyperparameters GB:", grid_search_gb.best_params_)

# ---- Tuning XGBoost ----
grid_search_xgb = GridSearchCV(estimator=xgb.XGBClassifier(random_state=42),
                               param_grid=param_grid_estimators_rate, cv=3, scoring='accuracy')
grid_search_xgb.fit(X_train, y_train)
best_xgb = grid_search_xgb.best_estimator_
print("Best Hyperparameters XGB:", grid_search_xgb.best_params_)

# ---- Tuning LightGBM ----
grid_search_lgb = GridSearchCV(estimator=lgb.LGBMClassifier(random_state=42),
                               param_grid=param_grid_estimators_rate, cv=3, scoring='accuracy')
grid_search_lgb.fit(X_train, y_train)
best_lgb = grid_search_lgb.best_estimator_
print("Best Hyperparameters LGB:", grid_search_lgb.best_params_)

# ---- Tuning AdaBoost ----
grid_search_ada = GridSearchCV(estimator=AdaBoostClassifier(random_state=42),
                               param_grid=param_grid_estimators_rate, cv=3, scoring='accuracy')
grid_search_ada.fit(X_train, y_train)
best_ada = grid_search_ada.best_estimator_
print("Best Hyperparameters AdaBoost:", grid_search_ada.best_params_)


```

Compare the performance of the models tuned to the best parameters. What is a parameter grid?
How does the GridSearch library choose the best estimator?
Which models perform best in this dataset? Do you have an intuition why?



```python
# Compare performance of tuned models
models = {'Gradient Boosting': best_gb, 'XGBoost': best_xgb, 'LightGBM': best_lgb, 'AdaBoost': best_ada}
for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"{name} Classification Accuracy:", metrics.accuracy_score(y_test, y_pred))
    print(f"Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print(f"Classification Report:\n", classification_report(y_test, y_pred))
```

## Extra: visualising AdaBoost

Visualise the estimators that the AdaBoost model is comprised of (you can get them using estimators_). Check the weights and errors of each of them (estimator_weights_ and estimator_errors_).
Tip: remember that the individual estimators are trees and we have already seen how to print trees in text or graph form.


```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt


# Access estimators, weights, and errors
estimators = best_ada.estimators_
estimator_weights = best_ada.estimator_weights_
estimator_errors = best_ada.estimator_errors_

# Visualize the first few estimators (decision trees)
for i, estimator in enumerate(estimators[:3]):  # Show first 3 trees for example
    plt.figure(figsize=(10, 5))
    plot_tree(estimator, filled=True)
    plt.title(f"Decision Tree Estimator {i+1}")
    plt.show()

# Check estimator weights and errors
print("Estimator Weights:", estimator_weights)
print("Estimator Errors:", estimator_errors)
```
