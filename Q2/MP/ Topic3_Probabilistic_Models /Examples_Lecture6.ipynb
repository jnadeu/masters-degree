## 1. Inference in Graphical Models


Inference in graphical models involves computing the marginal distribution of a subset of variables or 
conditional distributions given evidence. This can be visualized as propagating information through the graph.


```python

import matplotlib.pyplot as plt
import networkx as nx

G = nx.DiGraph()
edges = [("X1", "X2"), ("X2", "X3"), ("X3", "X4")]
G.add_edges_from(edges)

plt.figure(figsize=(5, 3))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=1000, font_size=12)
plt.title("Inference on a Chain")
plt.show()
```

### Example 1: Inference on a Tree Structure

```python

G = nx.DiGraph()
edges = [("X1", "X2"), ("X1", "X3"), ("X2", "X4"), ("X3", "X5")]
G.add_edges_from(edges)

plt.figure(figsize=(5, 3))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=1000, font_size=12)
plt.title("Inference on a Tree Structure")
plt.show()
```

### Example 2: Inference on a Loopy Graph

```python

G = nx.DiGraph()
edges = [("X1", "X2"), ("X2", "X3"), ("X3", "X1"), ("X2", "X4")]
G.add_edges_from(edges)

plt.figure(figsize=(5, 3))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=1000, font_size=12)
plt.title("Inference on a Loopy Graph")
plt.show()
```

## 2. Factor Graphs


Factor graphs are a way to represent the factorization of a function. Each variable and each factor are represented
as nodes, with edges connecting variables to their factors. Below is an example of a simple factor graph.


```python

G = nx.Graph()
G.add_nodes_from(["X1", "X2", "X3", "F1", "F2"])
edges = [("X1", "F1"), ("X2", "F1"), ("X2", "F2"), ("X3", "F2")]
G.add_edges_from(edges)

plt.figure(figsize=(5, 3))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=1000, font_size=12)
plt.title("Factor Graph Representation")
plt.show()
```

### Example 1: Directed Factor Graph

```python

G = nx.DiGraph()
edges = [("X1", "F1"), ("X2", "F1"), ("F1", "X3")]
G.add_edges_from(edges)

plt.figure(figsize=(5, 3))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=1000, font_size=12)
plt.title("Directed Factor Graph Representation")
plt.show()
```

### Example 2: Undirected Factor Graph

```python

G = nx.Graph()
edges = [("X1", "F1"), ("X2", "F1"), ("X2", "F2"), ("X3", "F2"), ("X3", "F1")]
G.add_edges_from(edges)

plt.figure(figsize=(5, 3))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=1000, font_size=12)
plt.title("Undirected Factor Graph Representation")
plt.show()
```

## 3. Sum-Product Algorithm


The Sum-Product algorithm is used in factor graphs to compute marginal distributions efficiently. The algorithm
propagates messages across the graph structure, aggregating information along the way.


```python

import numpy as np

# Define factors
f1 = np.array([[0.5, 0.8], [0.2, 0.1]])
f2 = np.array([0.6, 0.4])

# Sum-Product message passing
m1 = np.sum(f1 * f2.reshape(-1, 1), axis=0)
print("Message from Factor 1 to Variable X2:", m1)
```

### Example 1: Variable Elimination in Sum-Product

```python

m2 = np.sum(f1, axis=0)
print("Message after Variable Elimination:", m2)
```

### Example 2: Complex Propagation in Sum-Product

```python

m3 = np.sum(f1 * f2.reshape(-1, 1), axis=1)
print("Complex Propagation Message:", m3)
```
